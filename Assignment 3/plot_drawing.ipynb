```python
import matplotlib.pyplot as plt

# 编译优化选项
optimization_options = ["-O0", "-O1", "-O2"]

# gcc 和 clang 运行时间
gcc_runtimes = [18.78, 18.97, 19.61]
clang_runtimes = [8.42, 18.50, 8.45]

# 创建横坐标序列
x = range(len(optimization_options))

# 绘制 gcc 折线
plt.plot(x, gcc_runtimes, marker='o', label='gcc', linestyle='-')

# 绘制 clang 折线
plt.plot(x, clang_runtimes, marker='o', label='clang', linestyle='-')

# 设置 x 轴标签
plt.xticks(x, optimization_options)

# 设置图例
plt.legend()

# 设置标题和标签
plt.title("Running Time Comparison")
plt.xlabel("optimization options")
plt.ylabel("running time (s)")

# 显示图表
plt.grid(True)
plt.show()
```
