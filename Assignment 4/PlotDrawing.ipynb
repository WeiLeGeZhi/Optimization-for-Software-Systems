```python
import matplotlib.pyplot as plt

# 数据
labels = ['No Vectorization', 'Vectorization', 'AVX2 Optimization']
total_runtimes = [0.099, 0.034, 0.019]
user_runtimes = [0.098, 0.030, 0.015]
kernel_runtimes = [0.001, 0.005, 0.004]

# 宽度
bar_width = 0.35
index = range(len(labels))

# 绘制图表
fig, ax = plt.subplots()

bar1 = ax.bar(index, total_runtimes, bar_width, label='Total Runtime')
bar2 = ax.bar([i + bar_width for i in index], user_runtimes, bar_width, label='User Runtime')
bar3 = ax.bar([i + 2 * bar_width for i in index], kernel_runtimes, bar_width, label='Kernel Runtime')

# 添加标签、标题和图例
ax.set_xlabel('Optimization Technique')
ax.set_ylabel('Time (seconds)')
ax.set_title('Performance Comparison')
ax.set_xticks([i + bar_width for i in index])
ax.set_xticklabels(labels)
ax.legend()

# 显示图表
plt.show()
```
