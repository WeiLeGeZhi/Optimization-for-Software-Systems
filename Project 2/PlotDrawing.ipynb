```python
import matplotlib.pyplot as plt
```

```python
# 参数
parameters = [5, 15, 30, 60, 90]

# loop.aarch64.clang的运行时间
runtime_clang = [0.885947, 2.615552, 5.133838, 9.790129, 15.235213]

# loop.aarch64.gcc的运行时间
runtime_gcc = [2.002463, 6.260712, 12.119254, 23.839527, 36.315605]

# 设置图表的宽度
width = 0.35

# 生成x轴坐标
x = range(len(parameters))

# 创建柱状图
fig, ax = plt.subplots()
bar1 = ax.bar(x, runtime_clang, width, label='loop.aarch64.clang')
bar2 = ax.bar([i + width for i in x], runtime_gcc, width, label='loop.aarch64.gcc')

# 设置图表的标签和标题
ax.set_xlabel('Input')
ax.set_ylabel('Running Time')
ax.set_title('Running Time of Different Compiler under qemu-aarch64')
ax.set_xticks([i + width / 2 for i in x])
ax.set_xticklabels(parameters)
ax.legend()

# 添加数值标签
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', (bar.get_x() + bar.get_width() / 2, height), ha='center', va='bottom')

add_labels(bar1)
add_labels(bar2)

# 显示图表
plt.show()
```

```python
# 参数
parameters = [5, 15, 30, 60, 90]

# 运行时间数据
runtime_clang_aarch64 = [0.885947, 2.615552, 5.133838, 9.790129, 15.235213]
runtime_gcc_aarch64 = [2.002463, 6.260712, 12.119254, 23.839527, 36.315605]
runtime_clang_x64 = [0.153707, 0.454337, 0.926167, 1.827817, 2.698675]
runtime_gcc_x64 = [0.588919, 1.736898, 3.271416, 6.554815, 9.469506]

# 设置图表的宽度
width = 0.2

# 生成x轴坐标
x = range(len(parameters))

# 创建柱状图
fig, ax = plt.subplots()
bar1 = ax.bar([i - 1.5 * width for i in x], runtime_clang_aarch64, width, label='loop.aarch64.clang')
bar2 = ax.bar([i - 0.5 * width for i in x], runtime_gcc_aarch64, width, label='loop.aarch64.gcc')
bar3 = ax.bar([i + 0.5 * width for i in x], runtime_clang_x64, width, label='loop.x64.clang')
bar4 = ax.bar([i + 1.5 * width for i in x], runtime_gcc_x64, width, label='loop.x64.gcc')

# 设置图表的标签和标题
ax.set_xlabel('Input')
ax.set_ylabel('Running Time')
ax.set_title('Running Time of Different Compiler under Different Environment')
ax.set_xticks([i for i in x])
ax.set_xticklabels(parameters)
ax.legend()

# 添加数值标签
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', (bar.get_x() + bar.get_width() / 2, height), ha='center', va='bottom')

add_labels(bar1)
add_labels(bar2)
add_labels(bar3)
add_labels(bar4)

# 显示图表
plt.show()
```
